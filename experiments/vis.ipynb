```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to your log file
log_file_path = 'experiments/runs/multiple_lora/2024-02-16T05-35-03_multiple_lora/eval_metrics.txt'
# Read the log file into a pandas DataFrame
df = pd.read_csv(log_file_path)
```

```python
val_loss = df[df['val_task'] == 0][df['task'] == 0][df['metric_name']=='val_loss'][~df['step'].isna()]

# Plotting
plt.figure(figsize=(10, 6))

# Plot validation metric
plt.plot(val_loss['metric_value'], label='Validation Metric', linestyle='--')

plt.xlabel('Epoch')
plt.ylabel('Metric Value')
plt.title('Validation Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
